```python
import pandas as pd
import matplotlib.pyplot as plt
file_path = "cell_phones_total.csv"
data = pd.read_csv(file_path)
def convert_shorthand(value):
    if isinstance(value, str):
        value = value.replace(",", "")  # Remove commas if present
        if "k" in value:
            return float(value.replace("k", "")) * 1e3
        elif "M" in value:
            return float(value.replace("M", "")) * 1e6
    try:
        return float(value)
    except ValueError:
        return None  
years = data.columns[1:]
for year in years:
    data[year] = data[year].apply(convert_shorthand)
global_trend = data[years].sum()
plt.figure(figsize=(8, 10))  
plt.plot(global_trend.values, years, marker="o", linestyle="-", linewidth=2.5, markersize=6, color="darkblue", label="Total Cell Phones")
plt.ylabel("Year", fontsize=12, fontweight="bold")  # Year on Y-axis
plt.xlabel("Total Number of Cell Phones", fontsize=12, fontweight="bold")
plt.title("Global Rise in Cell Phone Usage Over the Years", fontsize=14, fontweight="bold", color="darkblue")
plt.gca().invert_yaxis()
plt.grid(True, linestyle="--", alpha=0.6)
plt.tight_layout()
plt.show()
```
