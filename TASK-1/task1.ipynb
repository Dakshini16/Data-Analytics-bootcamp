```python
import pandas as pd
import matplotlib.pyplot as plt
file_path = "lex.csv"  
df = pd.read_csv(file_path)
df_numeric = df.drop(columns=['country'])
avg_life_expectancy = df_numeric.mean()
years = avg_life_expectancy.index.astype(int)
life_expectancy = avg_life_expectancy.values
plt.figure(figsize=(10, 6))
plt.plot(life_expectancy, years, marker='o', linestyle='-', color='r')
plt.xlabel("Average Life Expectancy")
plt.ylabel("Year")
plt.title("Average Life Expectancy Over Time")
plt.grid(True)
plt.show()
```
