```python
import pandas as pd
import string
import matplotlib.pyplot as plt
from wordcloud import WordCloud

# Load the CSV file
file_path = "IPL_Match_Highlights_Commentary.csv"
df = pd.read_csv(file_path)

# Extract commentary text
text_data = " ".join(df["Commentary"].dropna())

# Define a set of common stopwords manually
common_stopwords = {
    "to", "the", "a", "of", "and", "in", "for", "on", "with", "at", "by", "from", 
    "is", "it", "that", "this", "as", "an", "be", "are", "was", "were", "or", "so", 
    "if", "not", "but", "he", "she", "they", "you", "we", "his", "her", "their", 
    "our", "its", "about", "can", "out", "up", "down", "over", "under", "then", "than",
    "all", "any", "some", "just", "one", "two", "there", "here", "what", "which", 
    "who", "whom", "how", "when", "where", "why", "have", "has", "had", "do", "does", 
    "did", "me", "my", "mine", "your", "yours", "him", "them", "these", "those", 
    "been", "being", "will", "would", "should", "could", "shall", "may", "might", 
    "must", "now"
}

# Tokenize and clean text
words = text_data.split()
filtered_words = [word.lower().strip(string.punctuation) for word in words if word.lower() not in common_stopwords]

# Generate word cloud
wordcloud = WordCloud(width=800, height=400, background_color="white", colormap="viridis").generate(" ".join(filtered_words))

# Plot word cloud
plt.figure(figsize=(10, 5))
plt.imshow(wordcloud, interpolation="bilinear")
plt.axis("off")
plt.show()
```

```python
import pandas as pd
import string
import matplotlib.pyplot as plt
from wordcloud import WordCloud
import seaborn as sns
file_path = "IPL_Match_Highlights_Commentary.csv"
df = pd.read_csv(file_path)
text_data = " ".join(df["Commentary"].dropna())
common_stopwords = {
    "to", "the", "a", "of", "and", "in", "for", "on", "with", "at", "by", "from", 
    "is", "it", "that", "this", "as", "an", "be", "are", "was", "were", "or", "so", 
    "if", "not", "but", "he", "she", "they", "you", "we", "his", "her", "their", 
    "our", "its", "about", "can", "out", "up", "down", "over", "under", "then", "than",
    "all", "any", "some", "just", "one", "two", "there", "here", "what", "which", 
    "who", "whom", "how", "when", "where", "why", "have", "has", "had", "do", "does", 
    "did", "me", "my", "mine", "your", "yours", "him", "them", "these", "those", 
    "been", "being", "will", "would", "should", "could", "shall", "may", "might", 
    "must", "now"
}
words = text_data.split()
filtered_words = [word.lower().strip(string.punctuation) for word in words if word.lower() not in common_stopwords]
wordcloud = WordCloud(
    width=1000, height=500, 
    background_color="black", 
    colormap="coolwarm", 
    contour_color="white", 
    contour_width=2, 
    max_words=200, 
    font_path=None
).generate(" ".join(filtered_words))
plt.figure(figsize=(12, 6))
sns.set_style("dark")
plt.imshow(wordcloud, interpolation="bilinear")
plt.axis("off")
plt.title(" IPL Match Commentary - Word Cloud", fontsize=18, fontweight="bold", color="white")
plt.gca().set_facecolor("#2C3E50")
plt.show()
```
