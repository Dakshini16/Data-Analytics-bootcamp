```python
import pandas as pd
import matplotlib.pyplot as plt
file_path = "mincpcap_cppp.csv"
df_income = pd.read_csv(file_path)
df_income_numeric = df_income.drop(columns=['country'])
avg_income = df_income_numeric.mean()
years = avg_income.index.astype(int)  # Convert index to int (years)
income_values = avg_income.values  # Corresponding income values
plt.figure(figsize=(10, 6))
plt.plot(income_values, years, marker='o', linestyle='-', color='royalblue', 
         linewidth=2.5, markersize=7, markerfacecolor='red', markeredgecolor='black')
for i in range(0, len(years), max(1, len(years) // 10)):  # Label every ~10% of points
    plt.text(income_values[i], years[i], f"{income_values[i]:,.0f}", fontsize=9, 
             ha='right', va='bottom', color="black")
plt.xlabel("Average Income (per capita)", fontsize=12, fontweight="bold")
plt.ylabel("Year", fontsize=12, fontweight="bold")
plt.title("Average Income Over Time", fontsize=14, fontweight="bold", color="darkblue")
plt.grid(True, linestyle="--", alpha=0.5)
plt.show()
```
